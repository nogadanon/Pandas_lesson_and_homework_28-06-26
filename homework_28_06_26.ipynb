```python
import pandas as pd
```

Load tips.csv. Display the first rows, df.info(), and df.describe().T. Then add a new column price_per_person calculated as total_bill / size.

```python
tips_df = pd.read_csv('tips.csv')
display(tips_df.head())
display(tips_df.info())
display(tips_df.describe().T)
tips_df['price_per_person'] = tips_df['total_bill']/tips_df['size']
tips_df
```

Count how many orders (rows) were placed at each time and for each gender.

```python
tips_df.groupby(['time', 'sex']).size().to_frame()
```

Count how many orders (rows) were placed on each day.

```python
tips_df.groupby('day').size()
```

Find the maximum price_per_person for each day and gender combination.

```python
tips_df.groupby(['day', 'sex'])[['price_per_person']].max()
```

Find the minimum price_per_person for each day and size combination.

```python
tips_df.groupby(['day', 'size'])[['price_per_person']].min()
```

Calculate the sum of total_bill grouped by size and smoker status. Then use .loc to extract the sum for size 2 smokers only.

```python
tips_df.groupby(['size', 'smoker'])[['total_bill']].sum()
```

Display the maximum price_per_person grouped by day and gender. Then use .loc to show the results for Male only across all days.

```python
max_price_df = tips_df.groupby(['day', 'sex'])[['price_per_person']].max()
display(max_price_df)
max_price_df.swaplevel().loc['Male']
```
